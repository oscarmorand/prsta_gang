```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
tornadoes_dataset_path = "tornado_data.csv"
tornadoes_data = pd.read_csv(tornadoes_dataset_path)
```

```python
print(tornadoes_data.head())
```

```python
months = tornadoes_data.columns[1:]

plt.figure(figsize=(12, 6))
for year in tornadoes_data["Year"].unique():
    year_data = tornadoes_data[tornadoes_data["Year"] == year]
    plt.plot(months, year_data.values[0][1:], label=year, alpha=0.5)
plt.show()
```

```python
# agglomerate all the year to see the mean per month

mean_per_month = tornadoes_data[months].mean()
std_per_month = tornadoes_data[months].std()

plt.figure(figsize=(12, 6))
plt.plot(months, mean_per_month, label="Mean")
plt.fill_between(months, mean_per_month - std_per_month, mean_per_month + std_per_month, alpha=0.2)
plt.show()
```

```python
# agglomerate all the months into a single value

tornadoes_data["Total"] = tornadoes_data.sum(axis=1)

plt.plot(tornadoes_data["Year"], tornadoes_data["Total"])
```

```python
tornadoes_raw_data = tornadoes_data.values[:, 1:-1].flatten()

plt.figure(figsize=(12, 6))
plt.plot(tornadoes_raw_data)
plt.show()
```

```python
episodes_dataset_path = "nino_nina_prsta_month.csv"
episodes_data = pd.read_csv(episodes_dataset_path)

print(episodes_data.head())
```

```python
print(np.isclose(episodes_data["TOTAL"].values, (episodes_data["ClimAdjust"].values + episodes_data["ANOM"].values), atol=1e-2).all())
```

```python
x = episodes_data["YR"].astype(str) + '-' + episodes_data["MON"].astype(str)

plt.figure(figsize=(12, 6))
plt.fill_between(range(len(x)), episodes_data["ANOM"], step="pre", alpha=0.3)
plt.plot(episodes_data["ANOM"], color='black', linewidth=0.7)
step = 24
plt.xticks(ticks=range(0, len(x), step), labels=x[::step], rotation=45, alpha=0.4)
for i in range(-4, 5):
    if i == 0:
        continue
    j = i / 2
    color = "red" if i > 0 else "blue"
    plt.plot(np.ones(len(x)) * j, color=color, linestyle="--")
plt.show()
```

```python
episodes_data['RollingMean'] = episodes_data['ANOM'].rolling(window=3, center=True).mean()

plt.figure(figsize=(12, 6))
plt.fill_between(range(len(x)), episodes_data["RollingMean"], step="pre", alpha=0.3)
plt.plot(episodes_data["RollingMean"], color='black', linewidth=0.7)
step = 24
plt.xticks(ticks=range(0, len(x), step), labels=x[::step], rotation=45, alpha=0.4)
for i in range(-4, 5):
    if i == 0:
        continue
    j = i / 2
    color = "red" if i > 0 else "blue"
    plt.plot(np.ones(len(x)) * j, color=color, linestyle="--")
plt.xlabel("Year-Month")
plt.ylabel("Anomaly")
plt.title("ENSO Anomaly")
plt.show()
```

```python
episodes_data.loc[0, "RollingMean"] = (episodes_data.loc[0, "ANOM"] + episodes_data.loc[1, "ANOM"]) / 2
episodes_data.drop([len(x) -1], inplace=True)

print(episodes_data.head())
print(episodes_data.tail(20))
```

```python
episodes_data['special'] = episodes_data['RollingMean'].apply(lambda x: 1 if x >= 0.5 else -1 if x <= -0.5 else 0)
x = x[:-1]

print(episodes_data.head())
```

```python
plt.figure(figsize=(12, 6))
plt.plot(x, episodes_data["special"])
plt.xticks(ticks=range(0, len(x), step), labels=x[::step], rotation=45, alpha=0.4)
plt.show()
```

```python
episodes_data['StreakLength'] = episodes_data.groupby((episodes_data['special'] != episodes_data['special'].shift()).cumsum())['special'].transform('size')

# Set streak length to 0 where the value is 0 (optional, based on your requirements)
episodes_data.loc[episodes_data['special'] == 0, 'StreakLength'] = 0

episodes_data['StreakFlag'] = episodes_data['StreakLength'] >= 5

episodes_data['StreakFlag'] = episodes_data.apply(
    lambda row: -1 if row['special'] == -1 and row['StreakLength'] >= 5
    else (1 if row['special'] == 1 and row['StreakLength'] >= 5 else 0),
    axis=1
)

episodes_data.drop(columns="StreakLength", inplace=True)
episodes_data.drop(columns="special", inplace=True)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(x, episodes_data["StreakFlag"])
plt.xticks(ticks=range(0, len(x), step), labels=x[::step], rotation=45, alpha=0.4)
plt.yticks([-1, 0, 1], ["La Nina", "Neutral", "El Nino"])
plt.xlabel("Year-Month")
plt.ylabel("ENSO Episode")
plt.title("ENSO Episodes")
plt.show()
```

```python
signal = episodes_data["StreakFlag"].values

episodes_lengths = []
prec = 0
in_episode = False
counter = 0
for i in range(len(signal)):
    value = signal[i]
    if prec == 0:
        if value != 0:
            in_episode = True
    else:
        if value == prec:
            counter += 1
        else:
            in_episode = False
            episodes_lengths.append(counter)
            counter = 1
    prec = value

print(episodes_lengths)

plt.hist(episodes_lengths, bins=range(1, 50, 2))
plt.show()
    
```

```python
from scipy import stats

el_nino = acres_nino
el_nina = acres_nina

shapiro_nino = stats.shapiro(el_nino)
shapiro_nina = stats.shapiro(el_nina)
print(f"Shapiro El Niño: p-value = {shapiro_nino.pvalue}")
print(f"Shapiro La Niña: p-value = {shapiro_nina.pvalue}")

# Si les deux groupes sont normaux, vérifier l'égalité des variances
if shapiro_nino.pvalue > 0.05 and shapiro_nina.pvalue > 0.05:
    levene_test = stats.levene(el_nino, el_nina)
    print(f"Levene test (égalité des variances) : p-value = {levene_test.pvalue}")
    
    # Choix du test en fonction des variances
    if levene_test.pvalue > 0.05:
        t_test = stats.ttest_ind(el_nino, el_nina, equal_var=True)
    else:
        t_test = stats.ttest_ind(el_nino, el_nina, equal_var=False)
    print(f"T-test: p-value = {t_test.pvalue}")
else:
    # Si les données ne sont pas normales, effectuer un test non paramétrique
    mann_whitney = stats.mannwhitneyu(el_nino, el_nina, alternative='greater')
    print(f"Mann-Whitney U test: p-value = {mann_whitney.pvalue}")
```
